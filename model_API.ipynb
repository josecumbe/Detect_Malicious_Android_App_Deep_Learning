```python
app = Flask(__name__)
model = pickle.load(open('rf_model.pkl', 'rb'))


run_with_ngrok(app)
@app.route('/')
def home():
    return render_template('index.html')

@app.route('/getprediction',methods=['POST'])
def getprediction():    

    input = [float(x) for x in request.form.values()]
    final_input = [np.array(input)]
    prediction = model.predict(final_input)
    if (prediction == 0):
      prediction_value = 'Genuine/Benign'
    else:
      prediction_value = 'Fraudulent/Malicious'


    return render_template('results.html', output='Based on the network layer features values entered, there is a 73% chance the app is {}'.format(prediction_value))
   

if __name__ == "__main__":
    app.run()
```
